# Imports

```python
import pandas as pd
import os
```

Data sourced from https://www.niftyindices.com/reports/historical-data

# Concatenating all csv files to obtain all data in one file:

```python
ll = [f for f in os.listdir() if f.endswith(".csv")]
ll
```

```python
ll.sort()
ll
```

```python
df_list = []
for filename in ll:
    df = pd.read_csv(filename)
    df_list.append(df)

merged_df = pd.concat(df_list, ignore_index = True)
```

```python
merged_df
```

```python
merged_df.to_csv("NIFTY 50 24-05-2015 to 24-05-2024", index = False)
```

# Converting values in the Date column to numeric values

```python
df = pd.read_csv("NIFTY 50 Historical 01-01-2000 to 31-05-2024.csv")
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='mixed')
df
```

```python
type(df['Date'][0])
```

```python
cols = list(df.columns)
print(cols)
for col in cols:
    print(f"\nColumn: {col}")
    print(f"dtype: {df[col].dtype}")
    print(f"All vals are not null: {df[col].notnull().all()}")
```

```python
df.to_csv("NIFTY 50 2000-01-03 to 2024-05-31.csv", index = False)
```
